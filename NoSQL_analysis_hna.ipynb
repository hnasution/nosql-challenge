```python
# Import dependencies
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Number of {count} establishments with hygiene score of {query['scores.Hygiene']}.")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
print(f"There are {len(df)} entries shwon in this DataFrame.")

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': "London"},'RatingValue': {'$gte': 4}}
results = list(db.establishments.find(query))

# Use count_documents to display the number of documents in the result
print(f"There are {len(results)} establishments in London with a RatingValue greater than or equal to 4.\n")

# Display the first document in the results using pprint
pprint(results[0:1])
```

```python
# Convert the result to a Pandas DataFrame
convert_df = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
# print(convert_df)
print(f"Number of rows in this DataFrame: {len(convert_df)} rows.")

# Display the first 10 rows of the DataFrame
convert_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

# Find Rating value equal to 5
# Sort by hygiene score

query = {'RatingValue': 5,'geocode.latitude': {'$gte': latitude*(1-degree_search),'$lte': latitude*(1+degree_search)},\
         'geocode.longitude': {'$gte': longitude*(1-degree_search),'$lte': longitude*(1+degree_search)}}
fields = {'RatingValue': 1, 'BusinessName': 1,'geocode.latitude':1,'geocode.longitude':1}
sort = [('scores.Hygiene', 1)]

# Capture the results to a variable
results = list(db.establishments.find(query,fields).sort(sort))

# Use count_documents to display the number of documents in the result
print(f'the number of documents found is {len(results)}')

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
rating5_df = pd.json_normalize(results)
rating5_df.head(5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = { '$match' : {'scores.Hygiene': 0} }

# 2. Groups the matches by Local Authority
group_query= {'$group': {'_id': {"local Authority": "$LocalAuthorityName"},
                        'count': { '$sum':1} } }

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f'the number of documents found is {len(list(results))}')

# Print the first 10 results
# print(results[0:10])
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print(hygiene_df.count())

# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```
